```python
import csv

ljspeech_metafile = '/home/scpark/ai/datasets/LJSpeech-1.1/metadata.csv'
ljspeech_wavdir = '/home/scpark/ai/datasets/LJSpeech-1.1/wavs/'
lj_datasets = []
with open(ljspeech_metafile, newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter='|')
    for row in reader:
        path = ljspeech_wavdir + row[0] + '.wav'
        if '|' in row[1]:
            row[1] = row[1].split('|')[0]
            print(row[1])
        lj_datasets.append({'path': path, 'text': row[1], 'speaker': 0})
        
print('done')
```

```python
for data in lj_datasets[:100]:
    print(data)
```

```python
kss_dir = '/home/scpark/ai/datasets/kss'

kss_datasets = []
with open(kss_dir + '/transcript.v.1.3.txt') as f:
    while(True):
        line = f.readline()
        
        if not line:
            break
        
        line = line.split('|')
        path = kss_dir + '/' + line[0]
        text = line[2]
        
        kss_datasets.append({'path': path, 'text': text, 'speaker': 1})
            
print(len(kss_datasets))
```

```python
for data in kss_datasets[:100]:
    print(data)
```

```python
sm_data_dirs = ['/home/scpark/ai/datasets/tts_datasets/jeon',
             '/home/scpark/ai/datasets/tts_datasets/sunny',
             '/home/scpark/ai/datasets/tts_datasets/dio',
             '/home/scpark/ai/datasets/tts_datasets/yoona',
             '/home/scpark/ai/datasets/tts_datasets/suho',
             '/home/scpark/ai/datasets/tts_datasets/changmin',
             '/home/scpark/ai/datasets/tts_datasets/chanyeol',
             '/home/scpark/ai/datasets/tts_datasets/leeteuk']
```

```python
import json
from tqdm import tqdm

start_tokens = ['/audio']
sm_datasets = []
for index, data_dir in tqdm(enumerate(sm_data_dirs)):
    speaker = index + 2
    json_file = data_dir + '/alignment.json'
    with open(json_file) as f:
        data = json.load(f)
        for d in data:
            for start_token in start_tokens:
                start_index = d.find(start_token)
                if start_index >= 0:
                    break
                    
            sm_datasets.append({'path': data_dir + d[start_index:], 'text': data[d], 'speaker': index})
            
print('done')

for data in sm_datasets[:100]:
    print(data)
    print(data['text'].encode('utf-8'))
```

```python
data_files = lj_datasets + kss_datasets + sm_datasets
print(len(data_files))
```

```python
import os
import librosa

train_ratio, valid_ratio, test_ratio = 1.0, 0.0, 0.0
train_index = int(len(data_files) * train_ratio)
valid_index = int(len(data_files) * valid_ratio)
test_index = int(len(data_files) * test_ratio)
print(train_index, valid_index, test_index)

def write_filelist(filename, data_files):
    from tqdm.notebook import tqdm
    
    with open(filename, 'w') as file:

        for index, data_file in tqdm(enumerate(data_files)):
            
            sr = librosa.core.get_samplerate(data_file['path'])    
            data, sampling_rate = librosa.core.load(data_file['path'], sr=sr)
            if len(data) > sampling_rate * 10:
                print(len(data))
                continue
                
            l = data_file['path'] + '|' + data_file['text'] + '|' + str(data_file['speaker'])
            file.write(l + os.linesep)

write_filelist('filelists/lj+kss+sm_audio_text_train_filelist.txt', data_files[:train_index])
```
