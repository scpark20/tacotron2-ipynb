### LJ

```python
import csv

ljspeech_metafile = '/home/scpark/ai/datasets/LJSpeech-1.1/metadata.csv'
ljspeech_wavdir = '/home/scpark/ai/datasets/LJSpeech-1.1/wavs/'
lj_datasets = []
with open(ljspeech_metafile, newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter='|')
    for row in reader:
        path = ljspeech_wavdir + row[0] + '.wav'
        if '|' in row[1]:
            row[1] = row[1].split('|')[0]

        lj_datasets.append({'path': path, 'text': row[1], 'speaker': 0, 'language': 0, 'name': 'LJ'})
                 
for data in lj_datasets[:100]:
    print(data)
```

### libri5

```python
libri_speakers = []
with open('/home/scpark/ai/datasets/LibriTTS/SPEAKERS.txt') as f:
    while(True):
        line = f.readline()
        if len(line) <= 0:
            break
        if line.startswith(';'):
            continue
        line = line.split()

        id = line[0]
        gender = line[2]
        subset = line[4]
        duration = line[6]
        name = line[8]
        
        if gender == 'F' and subset == 'train-clean-360':
            libri_speakers.append({'id': id, 'duration': duration, 'name': name})
        
libri_speakers.sort(key=lambda x: float(x['duration']), reverse=True)
for speaker in libri_speakers[:10]:
    print(speaker)
```

```python
from os import listdir

libri_datasets = []
root_dir = '/home/scpark/ai/datasets/LibriTTS/train-clean-360'

for speaker_index, speaker in enumerate(libri_speakers[:10]):
    speaker_dir = root_dir + '/' + str(speaker['id'])
    
    for sub_dir in listdir(speaker_dir):
        dir = speaker_dir + '/' + sub_dir
        for file in listdir(dir):
            if 'wav' in file:
                
                wav_file = dir + '/' + file
                txt_file = wav_file[:-4] + '.normalized.txt'
                
                with open(txt_file) as f:
                    txt = f.readline()
                    
                libri_datasets.append({'path': wav_file, 'text': txt, 'speaker': speaker_index + 1, 'language': 0, 'name': speaker['name']})
                
for data in libri_datasets[:100]:
    print(data)
    
print(libri_datasets[-1])
```

### kss

```python
kss_dir = '/home/scpark/ai/datasets/kss'

kss_datasets = []
with open(kss_dir + '/transcript.v.1.3.txt') as f:
    while(True):
        line = f.readline()
        
        if not line:
            break
        
        line = line.split('|')
        path = kss_dir + '/' + line[0]
        text = line[2]
        
        kss_datasets.append({'path': path, 'text': text, 'speaker': 11, 'language': 1, 'name': 'kss'})
            
for data in kss_datasets[:100]:
    print(data)
```

### sm3

```python
sm_data_dir = '/home/scpark/ai/datasets/tts_datasets/'
sm_names = ['jeon', 'sunny', 'yoona']
```

```python
import json
from tqdm import tqdm

start_tokens = ['/audio']
sm_datasets = []
for index, name in enumerate(sm_names):
    data_dir = sm_data_dir + name
    speaker = index + 12
    json_file = data_dir + '/alignment.json'
    with open(json_file) as f:
        data = json.load(f)
        for d in data:
            for start_token in start_tokens:
                start_index = d.find(start_token)
                if start_index >= 0:
                    break
                    
            sm_datasets.append({'path': data_dir + d[start_index:], 'text': data[d], 'speaker': speaker, 'language': 1, 'name': name})
            
for data in sm_datasets[:100]:
    print(data)
```

```python
datasets = lj_datasets + libri_datasets + kss_datasets + sm_datasets
print(len(datasets))

speaker = -1
for data in datasets:
    if data['speaker'] > speaker:
        print(data)
        speaker = data['speaker']
        
```

```python
import os
import librosa

train_ratio, valid_ratio, test_ratio = 1.0, 0.0, 0.0
train_index = int(len(datasets) * train_ratio)
valid_index = int(len(datasets) * valid_ratio)
test_index = int(len(datasets) * test_ratio)
print(train_index, valid_index, test_index)

def write_filelist(filename, data_files):
    from tqdm.notebook import tqdm
    
    with open(filename, 'w') as file:

        for index, data_file in tqdm(enumerate(data_files)):
            
            sr = librosa.core.get_samplerate(data_file['path'])    
            data, sampling_rate = librosa.core.load(data_file['path'], sr=sr)
            if len(data) > sampling_rate * 10:
                print(len(data))
                continue
                
            l = data_file['path'] + '|' + data_file['text'] + '|' + str(data_file['speaker']) + '|' + str(data_file['language']) + '|' + data_file['name']
            file.write(l + os.linesep)

write_filelist('filelists/lj+libri+kss+sm_audio_text_train_filelist.txt', datasets[:train_index])
write_filelist('filelists/lj+libri+kss+sm_audio_text_val_filelist.txt', datasets[train_index:valid_index])
write_filelist('filelists/lj+libri+kss+sm_audio_text_test_filelist.txt', datasets[valid_index:test_index])
```
